### The distance from Lagos to other cities in Nigeria

Data - worldcities data (Basic) https://simplemaps.com/data/world-cities

Using: `Geopy` to calculate the geodesic distance between coordinates. https://geopy.readthedocs.io/en/stable/#module-geopy.distance

```python
import os
import csv
from geopy import distance
```

```python
# Read the data (exploring its structure)

input_dir = 'data'
input_filename = 'worldcities.csv'
input_path = os.path.join(input_dir, input_filename)

with open(input_path, 'r', encoding = 'utf-8') as input:
    csv_reader = csv.DictReader(input, delimiter = ',', quotechar = '"')
    for index, row in enumerate(csv_reader):
        print(f'{row} \n')
        if index == 1:
            break
```

```python
output_dir = 'result'
output_filename = 'cities_distance_from_lagos.csv'
output_path = os.path.join(output_dir, output_filename)

if not os.path.exists(output_dir):
    os.mkdir(output_dir)
```

```python
country = 'Nigeria'
city = 'Lagos'

with open(output_path, 'w', newline = '', encoding = 'utf-8') as output:
    fieldnames = ['city', 'state', 'distance_from_lagos (km)']
    
    csv_writer  = csv.DictWriter(output, fieldnames)
    csv_writer.writeheader()
    
    with open(input_path, 'r', encoding = 'utf-8') as input:
        csv_reader = csv.DictReader(input, delimiter = ',', quotechar = '"')
        for row in csv_reader:
            if row['city_ascii'] == city:
                city_coordinates = row['lat'], row['lng']

            if row['country'] == country and row['city_ascii'] != city:
                other_cities = row['city_ascii']
                state = row['admin_name']

                other_city_coordinates = row['lat'], row['lng']
                distance_from_lagos = distance.geodesic(city_coordinates, other_city_coordinates)
                csv_writer.writerow({'city': other_cities, 'state': state, 'distance_from_lagos (km)': f'{distance_from_lagos.km:.2f}'})

                print(f'Lagos to {other_cities} ({state}) - {distance_from_lagos}')
```

```python
#view the file that was written and explore

file_dir = 'result'
filename = 'cities_distance_from_lagos.csv'
filepath = os.path.join(file_dir, filename)

with open(filepath, 'r', encoding = 'utf-8') as file:
    csv_reader = csv.DictReader(file, delimiter = ',', quotechar = '"')
    for index, row in enumerate(csv_reader):
        print(row)
        if index == 4:
            break
```

```python
#count how many cities are in the country (Nigeria) BASED ON THE DATA

with open(filepath, 'r', encoding = 'utf-8') as file:
    csv_reader = csv.DictReader(file, delimiter = ',', quotechar = '"')

    count = 0
    for row in csv_reader:
        count += 1
        
print(count)
```
